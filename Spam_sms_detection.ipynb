# import necessary libraries

- Spam SMS detection is a machine learning task aimed at automatically identifying and filtering out unwanted, unsolicited, or fraudulent messages (spam) from legitimate ones (ham)
- Spam: Unwanted messages that may include advertisements, scams, or malicious content.
- Ham: Legitimate messages from trusted contacts, services, or companies.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

# Read the dataset

```python
df = pd.read_csv("spam.csv",encoding='ISO-8859-1')
```

# sanity check for data

# data preprocessing

```python
df.rename(columns={"v1":"checker","v2":"sms"},inplace=True)
```

```python
df
```

```python
vectorizer = TfidfVectorizer()
```

```python
x = vectorizer.fit_transform(df['sms'])
```

```python
encoder = LabelEncoder()
```

```python
df["checker"] = encoder.fit_transform(df["checker"])
```

```python
df
```

```python
y = df["checker"]
```

# Naive Bayes model

```python
nv = MultinomialNB()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
nv.fit(x_train,y_train)
```

```python
nb_pred = nv.predict(x_test)
```

```python
print(classification_report(y_test,nb_pred))
```

# Logistic regression model

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr_pred = lr.predict(x_test)
```

```python
print(classification_report(y_test,lr_pred))
```

# Support vector machine model

```python
svm = SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
svm_pred = svm.predict(x_test)
```

```python
print(classification_report(y_test,svm_pred))
```
